```python
# Выведите на консоль значение площади этого круга с точностью до 4-х знаков после запятой
# Площадь круга - число пи умножить на радиус в квадрате,
# Точность указывается в функции round()

radius = 36    # значение радиуса круга
P = 3.1415926  # Число пи
print('Площадь круга - ', round((P * radius ** 2),4))
```
