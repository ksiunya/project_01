```python
import random

my_favorite_songs = {
    'Waste a Moment': 3.03,
    'New Salvation': 4.02,
    'Staying\' Alive': 3.40,
    'Out of Touch': 3.03,
    'A Sorta Fairytale': 5.28,
    'Easy': 4.15,
    'Beautiful Day': 4.04,
    'Nowhere to Run': 2.58,
    'In This World': 4.02,
}

random_song = random.choices(list(my_favorite_songs.items()), k=3)
print(random_song)

total_song_duration = list()

for song, duration in random_song:
  total_song_duration.append(duration)
print(total_song_duration) 

tsd = sum(total_song_duration)
print('Три песни звучат', round(tsd,2), 'минут')
```
